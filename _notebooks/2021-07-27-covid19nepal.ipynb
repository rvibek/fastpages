```python
import pandas as pd
import altair as alt

import plotly.express as px

source = 'https://raw.githubusercontent.com/rvibek/covid19_nepal/main/data/covid19aggregate.csv'
data = pd.read_csv(source)
```

```python
#collapse-show

data.tail(5)
```

```python
alt.Chart(data).mark_line().transform_fold(
    fold=['new_cases', 'avg7_new_cases'],
    as_=['category', 'y']
        ).encode(
    alt.X('date', axis=alt.Axis(labels=False)),
    alt.Y('y:Q'),
    alt.Color('category:N'),
    tooltip=['total_cases', 'avg7_new_cases', 'date'],
    
).properties(
    width=800,
    height=400
)
```
